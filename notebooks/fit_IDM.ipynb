```python
import sys 
import numpy as np
import matplotlib.pyplot as plt
sys.path.append('..')

from lya.data import get_data_transfer_function, plot_pk
from lya.analysis import *

data = get_data_transfer_function()
```

```python
from lya.analysis import objective

def log_prob(pars, data):
    if pars < 1e-40 or pars > 1e-10:
        return -np.inf
    
    lnp = -0.5 * objective(pars, data)
    if np.isnan(lnp):
        lnp = -np.inf
    return lnp
```

```python
sigmas = np.logspace(-40,-16,20)
sigmas
```

```python
lnps = [-log_prob(sigma, data) for sigma in sigmas]
```

```python
plt.scatter(sigmas, lnps,s=5)
plt.xscale('log')
plt.yscale('log')
plt.ylabel('-loglikeliho')
```
